```python
import matplotlib.pyplot as plt    

x=[1,2,3,4,5]
y=[2,3,5,6,2]
y1=[1,2,3,4,5]

plt.figure(figsize=(22,22))

plt.subplot(2,2,1)
plt.plot(x,y,marker='.',linestyle='--',linewidth=1)
plt.title('Plot 1')

plt.subplot(2,2,1)  ## both are same Plot number
plt.plot(y,x,marker='.')
plt.title('plot 2')
```

```python
## Multiple plots
## Sample Data

x=[1,2,3,4,5]
y1=[1,4,9,16,25]
y2=[1,2,3,4,5]

plt.figure(figsize=(9,9)) ## the size of graph paper

plt.subplot(2,2,1) ## 2,2 mean 2 row and 2 column and last 1 is the number of plot 
plt.plot(x,y1,color='green')
plt.title("Plot 1")


plt.subplot(2,2,2)
plt.plot(y1,x,color='orange')
plt.title("Plot2")

plt.subplot(2,2,3)
plt.plot(y1,x,color='blue')
plt.title("Plot3")

plt.subplot(2,2,4)
plt.plot(y1,x,color='black')
plt.title("Plot2")
```

Bar Plot


```python
## Bar Plot
categories=['A','B','C','D','E']
values=[5,7,3,8,6]

## Create a bar plot
plt.bar(categories,values,color='purple')

plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot')
plt.show()
```

Pie Chat

```python
# Pie Chat 

labels=['A','B','C','D']
sizes=[30,20,40,10]
colors=['gold','yellowgreen','lightcoral','lightskyblue']      
explode=(0.2,0,0,0)

## Create a pie chart
# plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct="%1.1f%%",shadow=True)
plt.pie(sizes,colors=colors,labels=labels,explode=explode,shadow=True,autopct="%1.1f%%")
```

```python
## sales Data visualization
import pandas as pd
sales_data=pd.read_csv('friends.csv')

sales_data.head(8)
 
```

```python
sales_data.info()
```

```python
## Plot total sales By products
total_sales_by_product=sales_data.groupby('Item Type')['Total Cost'].sum()

total_sales_by_product
```

```python
total_sales_by_product.plot(kind='bar',color='teal')
```

```python
## Plot sales tread over time

sales_trend=sales_data.groupby('Order Date')['Total Revenue'].sum().reset_index()

plt.plot(sales_trend['Order Date'],sales_trend['Total Revenue'])
```
