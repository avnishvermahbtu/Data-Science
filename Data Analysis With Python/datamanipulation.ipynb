Pandas GroupBy

Splitting Data into Groups

```python
# Data Aggregating And Grouping

df
```

```python
# Grouping data by sorting keys: Group keys are sorted by default using the groupby operation. 
# Now we apply groupby() using sort.

group_mean=df.groupby('City')['Age'].mean()
print(group_mean)
```

```python
# Grouping data with multiple keys
#  Here we group a data of "City" and "Name" together using multiple keys in groupby function.
 
group_sum=df.groupby(['City','Name'])['Score'].sum()
print(group_sum)
```

Applying Aggregation with GroupBy

```python
# 1. Aggregation: It allows us to find a summary statistic for each group like summing 
# or averaging values

grouped_agg=df.groupby('Name')['Score'].agg(['mean','sum','count'])
grouped_agg
```

```python
## Merging and joining Dataframes
## Create simple DataFrames

df1=pd.DataFrame({'key':['A','B','C'],'Name':[1,2,3]})
df3=pd.DataFrame({'key':['A','B','D'],'Name':[4,5,6]})

```

```python
df1
```

```python
df3
```

Pandas DataFrame merge() Method

```python
# Merge DataFrame on the 'Key Columns'
pd.merge(df1,df3,on="key",how="inner")
```

```python
pd.merge(df1,df3,on='key',how='right') ## more priorty on rigth data
```

```python
pd.merge(df1,df3,on='key',how='left')  ## more priorty on left data
```

```python
pd.merge(df1,df3,on='key',how='outer')
```
